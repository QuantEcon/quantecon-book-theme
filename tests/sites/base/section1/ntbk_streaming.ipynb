# Test Streaming Output

This notebook tests streaming output from multiple print statements.

```python
# Test with streaming output from loop
# Each print statement creates a separate stream output
for i in [25, 50, 75]:
    error = 100 / (i / 10)
    print(f"Error at iteration {i} is {error}.")
```

```python
# Test with multiple consecutive print statements
# Each creates a separate stream output
for i in range(1, 6):
    print(f"Line {i}")
```
